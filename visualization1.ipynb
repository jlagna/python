#sales trend over time

```python
sales_by_date=df.groupby('Order Date')['Sales'].sum()
sales_by_date.head()
```

```python
plt.plot(sales_by_date.index,sales_by_date.values)
plt.title("Daily Sales Trend")
plt.xticks(rotation=90)
plt.show()
#1st graph
```

```python
plt.figure(figsize=(20,5))
sales_by_date=df.groupby('Order Date')['Sales'].sum()
sales_by_date.head()
plt.plot(sales_by_date.index,sales_by_date.values,color='r',linestyle='--')
plt.title("Daily Sales Trend",fontsize=16)
plt.xlabel("Order Date",fontsize=12)
plt.ylabel("Sales",fontsize=12)
plt.xticks(rotation=90)
plt.grid(True,linestyle='--')
plt.show()
```

###daily profit tends

```python
sales_by_date=df.groupby('Region')['Sales'].sum()
plt.bar(sales_by_date.index,sales_by_date.values,color='r')
plt.title("Total sales by region",fontsize=16)
plt.xlabel("Region",fontsize=12)
plt.ylabel("total Sales",fontsize=12)
plt.xticks(rotation=90)
plt.grid(True,linestyle='--')
plt.show()
```

```python
sales_by_region=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.bar(sales_by_region.index,sales_by_region.values,color='r')
plt.title("Total sales by region",fontsize=16)
plt.xlabel("Region",fontsize=12)
plt.ylabel("total Sales",fontsize=12)
#plt.xticks(rotation=90)
#plt.grid(True,linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(sales_by_region.index,sales_by_region.values,color=['#1f77b4','#ff7f0e','#2ca02c','#d62728'])
plt.title("Total sales by region",fontsize=16)
plt.xlabel("Region",fontsize=12)
plt.ylabel("Total Sales",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.barh(sales_by_region.index,sales_by_region.values,color=['#1f77b4','#ff7f0e'])
plt.title("Total sales by region",fontsize=16)
plt.xlabel("Region",fontsize=12)
plt.show()
```

##stack bar graph

```python
sales_by_region=df.groupby('Region')['Sales'].sum()
profit_by_region=df.groupby('Region')['Profit'].sum()
plt.bar(sales_by_region.index,sales_by_region.values,color='r')
plt.bar(profit_by_region.index,profit_by_region.values,bottom=sales_by_region.values,color= 'b',label='Profit')
plt.title("Total sales by Region",fontsize=16)
plt.xlabel("Region",fontsize=12)
plt.ylabel("Total Sales",fontsize=12)
plt.legend()
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,6))
plt.bar(sales_by_region.index,sales_by_region.values,facecolor='w',edgecolor='blue')
plt.title("Total sales by region",fontsize=16)
plt.xlabel("Region",fontsize=12)
plt.ylabel("Total Sales",fontsize=12)
plt.show()
```

```python
plt.figure()
sales_cat = df.groupby('Category')['Sales'].sum()
plt.pie(sales_cat.values,labels=sales_cat.index)
plt.title("Sales share by category")
plt.show()
```

```python
plt.figure()
sales_cat = df.groupby('Category')['Sales'].sum()
colors = ['#1f77b4','#ff7f0e','#2ca02c'] #custom colors
explode=[0.00,0.02,0.05] #pull out slice slightly

plt.pie(sales_cat.values,labels=sales_cat.index,explode=explode,colors=colors)
plt.title("Sales share by category")
plt.show()
```

```python
plt.figure()
sales_cat = df.groupby('Category')['Sales'].sum()
colors = ['#1f77b4','#ff7f0e','#2ca02c'] #custom colors
plt.pie(sales_cat.values,labels=sales_cat.index,colors=colors,autopct='%1.2f%%')
plt.title("Sales share by category")
plt.show()
```
